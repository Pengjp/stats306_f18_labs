### Aesthetic mappings

Aesthetics define how we map variables to features of the plot. Common examples:
- Position (x, y coordinates)
- Color
- Fill (‘inside’ coloring)
- Shape
- Linetype
- Size

Note that aesthetic mappings are inherited from the default aesthetic but can also be applied or overwritten for specific pieces.


### Geometric objects

What marks do we want on the plot. Each `geom_` object will have certain aes arguments that it requires or can use. Examples:

- `geom_point` for scatter plots
- `geom_line` for line plots
- `geom_boxplot`
- `geom_histogram`
- `geom_hline`, `geom_vline`, `geom_abline`
- `geom_text`

### Other layer types

- `stat_` functions are usually similar to `geom_` functions, simply named to emphasize their statistical operations.
- `scale_` functions change how aesthetics transform variables into features. For instance, changing color schemes or which line and point types are used, switching to log scales or reversing direction, etc. Also often used to modify legends, but some legend modifications use the `guide_` functions.
- `facet_` subsetting to multiple plots.
- `coord_` changing the coordinate systems
- `theme_` changing the visuals of the plot.

```r
ggplot(data = diamonds, aes(x=color, fill=cut)) + 
    geom_bar() +
    labs(x='Color', y='Count')
```

```r
ggplot(data = diamonds, aes(x=color, fill=cut)) + 
    geom_bar(position='dodge') +
    labs(x='Color', y='Count')
```

```r
ggplot(data=diamonds, aes(x = cut, y = price)) + 
    geom_boxplot() +
    labs(x='Cut Level', y='Price($)') + 
    ggtitle('Boxplot of Price by Cut')

# What are the points above the line?
```

```r
ggplot(data=diamonds, aes(x=carat, y=price)) + 
    geom_point() +
    geom_vline(xintercept=2.3, color='cyan') +
    geom_hline(yintercept=12000, color='magenta', linetype='dashed')
```

### Facets

Facets let us easily create multiple plots based on subsets of the data.

```r
ggplot(data=diamonds, aes(x=x, y=price)) + 
    geom_point() +
    facet_wrap(~ cut, nrow=2) +
    labs(x='Width', y='Price($)') 
```

```r
ggplot(data=diamonds, aes(x=x, y=price)) + 
    geom_point() +
    facet_grid(color ~ cut) +
    labs(x='Width', y='Price($)') 
```

```r
options(repr.plot.width=9, repr.plot.height=4)

ggplot(data=diamonds, aes(x=x, y=price)) + 
    geom_point() +
    facet_grid(. ~ color, labeller=label_both) +
    labs(x='Width', y='Price($)') 

options(repr.plot.width=5, repr.plot.height=4)
```
